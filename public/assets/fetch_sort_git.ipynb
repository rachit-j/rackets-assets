```python
import requests
from datetime import datetime, timedelta, timezone
from urllib.parse import urlparse
from collections import defaultdict

def get_github_commits(username, token, days_back=30):
    """
    Fetches and processes GitHub commit data for the specified user.

    :param username: GitHub username
    :param token: GitHub personal access token
    :param days_back: Number of days back to count commits
    :return: None
    """

    headers = {
        'Accept': 'application/vnd.github.cloak-preview',
        'Authorization': f'token {token}' # comment out if ur not using a token
    }

    cutoff_date = datetime.now(timezone.utc) - timedelta(days=days_back)

    def extract_repo_name(url):
        parsed_url = urlparse(url)
        path_parts = parsed_url.path.split('/')
        return '/'.join(path_parts[1:3])

    def process_commits(commits_data):
        last_commit_message = None
        last_commit_date = None
        total_commits_last_30_days = 0

        for commit_date_str, commit_message in commits_data:
            try:
                commit_date = datetime.strptime(commit_date_str, "%Y-%m-%dT%H:%M:%S.%f%z")
            except ValueError:
                commit_date = datetime.strptime(commit_date_str, "%Y-%m-%dT%H:%M:%SZ")
                commit_date = commit_date.replace(tzinfo=timezone.utc)

            if last_commit_date is None or commit_date > last_commit_date:
                last_commit_date = commit_date
                last_commit_message = commit_message

            if commit_date > cutoff_date:
                total_commits_last_30_days += 1

        return last_commit_date, last_commit_message, total_commits_last_30_days

    repo_commits = defaultdict(list)

    repos_url = f"https://api.github.com/users/{username}/repos"
    repos_response = requests.get(repos_url, headers=headers)
    repos_data = repos_response.json()

    for repo in repos_data:
        commits_url = repo['commits_url'].replace('{/sha}', '')
        commits_response = requests.get(commits_url, headers=headers)
        commits_data = commits_response.json()

        if isinstance(commits_data, list):
            for commit in commits_data:
                commit_date_str = commit['commit']['author']['date']
                commit_message = commit['commit']['message']
                repo_commits[repo['full_name']].append((commit_date_str, commit_message))

    search_url = f"https://api.github.com/search/commits?q=author:{username}"
    search_response = requests.get(search_url, headers=headers)
    search_results = search_response.json()

    for item in search_results['items']:
        repo_full_name = extract_repo_name(item['html_url'])
        commit_date_str = item['commit']['author']['date']
        commit_message = item['commit']['message']
        repo_commits[repo_full_name].append((commit_date_str, commit_message))

    header = f"{'Repository Full Name':<40} {'Last Commit Date':<30} {'Last Commit Message':<50} {'Commits in Last 30 Days':<25}"
    divider = "-" * len(header)

    print(header)
    print(divider)

    for repo_full_name, commits in repo_commits.items():
        last_commit_date, last_commit_message, total_commits_last_30_days = process_commits(commits)
        row = f"{repo_full_name:<40} {last_commit_date:%Y-%m-%d %H:%M:%S} {last_commit_message[:47] + '...' if last_commit_message else 'No commit message':<50} {total_commits_last_30_days:<25}"
        print(row)

# Example usage:
get_github_commits("rachit-j", "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")
```

```python
import pandas as pd

def display_csv_as_html_table(username, directory):
    """
    Reads a CSV file for the given username and returns HTML code for a simple table displaying the data.

    :param username: GitHub username
    :param directory: Directory where the CSV file is located
    :return: HTML string representing the table
    """
    file_path = f"{directory}/{username}.csv"
    try:
        # Read the CSV file into a DataFrame
        data = pd.read_csv(file_path)

        # Convert the DataFrame to HTML table
        html_table = data.to_html(index=False)
        return html_table

    except FileNotFoundError:
        return "CSV file not found for the given username."

# Example usage:
html_table = display_csv_as_html_table("theoh32", "/Users/rachitjaiswal/Desktop/Github/Mortys Stuff/users")
print(html_table)
```
